```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.io import ascii

%matplotlib inline
```

```python
GX9p9 = ascii.read("GX9p9_FeLedge.res_0.dat")
J1817 = ascii.read("J1817_FeLedge.res_0.dat")
CygX1 = ascii.read("CygX1_FeLedge.res_0.dat")
```

```python
def plot_data(ax, data, mcol='r'):
    ax.errorbar(data['col1'], data['col3'], yerr=data['col4'], 
                marker='', ls='', capsize=1, color='0.5', alpha=0.5)
    ax.step(0.5*(data['col1']+data['col2']), data['col3'], color='k')
    ax.plot(data['col1'], data['col5'], color=mcol, lw=2, alpha=0.8)
    
    ax.set_xlim(17,18)
    ax.set_ylabel("Flux [phot/s/cm$^2$/$\AA$]", size=18)
    ax.set_xlabel("Wavelength [Angs]", size=18)
    return

def plot_res(ax, data, rcol='k'):
    ax.step(0.5*(data['col1']+data['col2']), 
            data['col6'], color=rcol, alpha=0.8);
    ax.axhline(0.0, color='0.5', ls='--')
    
    ax.set_xlim(17,18)    
    ax.set_ylabel(r"$\chi$", size=18)
    ax.set_xlabel("Wavelength [Angs]", size=18)
    return
```

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure( figsize=(8,4) )

gs = gridspec.GridSpec(2, 1, height_ratios=[2,1])
ax1 = plt.subplot(gs[0])
plot_data(ax1, GX9p9)
ax1.set_ylim(0.006,0.025)
ax1.xaxis.set_ticklabels('')

plt.text(17.05, 0.022, 'GX 9+9 (ObsId 11072)', color='k', fontsize=16)

ax2 = plt.subplot(gs[1])
plot_res(ax2, GX9p9)
ax2.set_ylim(-4.5,3.5)
ax2.yaxis.set_ticks([-4,-2,0,2])

ax1.tick_params(labelsize=12)
ax2.tick_params(labelsize=12)

gs.update(hspace=0.0)
```

```python
fig = plt.figure( figsize=(8,4) )

gs = gridspec.GridSpec(2, 1, height_ratios=[2,1])
ax1 = plt.subplot(gs[0])
plot_data(ax1, J1817)
ax1.set_ylim(0.0405,0.1)
ax1.xaxis.set_ticklabels('')

plt.text(17.05, 0.09, 'XTE J1817-330 (ObsId 6618)', color='k', fontsize=16)

ax2 = plt.subplot(gs[1])
plot_res(ax2, J1817)
ax2.set_ylim(-4.5,3.5)
ax2.yaxis.set_ticks([-4,-2,0,2])

ax1.tick_params(labelsize=12)
ax2.tick_params(labelsize=12)

gs.update(hspace=0.0)
```

```python
fig = plt.figure( figsize=(8,4) )

gs = gridspec.GridSpec(2, 1, height_ratios=[2,1])
ax1 = plt.subplot(gs[0])
plot_data(ax1, CygX1)
ax1.set_ylim(0.05,0.18)
ax1.xaxis.set_ticklabels('')

plt.text(17.05, 0.16, 'Cyg X-1 (ObsId 3407)', color='k', fontsize=16)

ax2 = plt.subplot(gs[1])
plot_res(ax2, CygX1)
ax2.set_ylim(-4.5,3.5)
ax2.yaxis.set_ticks([-4,-2,0,2])

ax1.tick_params(labelsize=12)
ax2.tick_params(labelsize=12)

gs.update(hspace=0.0)
```

## Put it all in one big plot

```python
fig = plt.figure( figsize=(18,4) )

gs = gridspec.GridSpec(2, 3, height_ratios=[2,1])

ax0 = plt.subplot(gs[0])
plot_data(ax0, GX9p9)
ax0.set_ylim(0.006,0.027)
ax0.xaxis.set_ticklabels('')
plt.text(17.05, 0.024, 'GX 9+9', color='k', fontsize=16)
plt.text(17.70, 0.024, 'ObsId 11072', color='0.2', fontsize=16)


ax1 = plt.subplot(gs[1])
plot_data(ax1, J1817)
ax1.set_ylim(0.0405,0.1)
ax1.xaxis.set_ticklabels('')
plt.text(17.05, 0.091, 'XTE J1817-330', color='k', fontsize=16)
plt.text(17.70, 0.091, 'ObsId 6618', color='0.2', fontsize=16)
ax1.set_ylabel('')

ax2 = plt.subplot(gs[2])
plot_data(ax2, CygX1)
ax2.set_ylim(0.05,0.18)
ax2.xaxis.set_ticklabels('')
plt.text(17.05, 0.16, 'Cyg X-1', color='k', fontsize=16)
plt.text(17.70, 0.16, 'ObsId 3407', color='0.2', fontsize=16)
ax2.set_ylabel('')

ax0.tick_params(labelsize=12)
ax1.tick_params(labelsize=12)
ax2.tick_params(labelsize=12)

ax3 = plt.subplot(gs[3])
plot_res(ax3, GX9p9)
ax3.set_ylim(-4.5,3.5)
ax3.yaxis.set_ticks([-4,-2,0,2])

ax4 = plt.subplot(gs[4])
plot_res(ax4, J1817)
ax4.set_ylim(-4.5,3.5)
ax4.yaxis.set_ticks([-4,-2,0,2])
ax4.set_ylabel('')

ax5 = plt.subplot(gs[5])
plot_res(ax5, CygX1)
ax5.set_ylim(-4.5,3.5)
ax5.yaxis.set_ticks([-4,-2,0,2])
ax5.set_ylabel('')

ax3.tick_params(labelsize=12)
ax4.tick_params(labelsize=12)
ax5.tick_params(labelsize=12)

gs.update(hspace=0.0)
gs.update(wspace=0.1)
```
